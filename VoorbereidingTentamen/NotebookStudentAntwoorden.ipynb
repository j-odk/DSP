```python
import sqlite3
import pandas as pd

def draai_query(query, database, return_data = True):
    with sqlite3.connect(database) as con:
        if return_data:
            res = pd.read_sql_query(query, con)
        else:
            con.execute(query)
            res=None
    return res

def lijst_tabellen(database):
    return draai_query("SELECT name FROM sqlite_master WHERE type='table';", database)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

### vraag 2a

```python
# voeg hier je antwoord op vraag 2a toe.

df_inventory_parts = pd.read_csv('vraag2/vraag2_inventory_parts.csv')
df_colors = pd.read_csv('vraag2/vraag2_colors.csv')
df_part_categories = pd.read_csv('vraag2/vraag2_part_categories.csv')
df_parts = pd.read_csv('vraag2/vraag2_parts.csv')

merged = pd.merge(
    left=df_inventory_parts, 
    right=df_colors, 
    how='left', 
    left_on='color_id', 
    right_on='id')

merged = pd.merge(
    left=merged, 
    right=df_parts, 
    how='left', 
    left_on='part_num', 
    right_on='part_num')

merged = pd.merge(
    left=merged, 
    right=df_part_categories, 
    how='left', 
    left_on='part_cat_id', 
    right_on='id')

# Make sure to use inplace=True else the column name will not be replaced.
merged.rename(columns={'name_x':'color_name'}, inplace=True)
merged.rename(columns={'name_y':'part_name'}, inplace=True)
merged.rename(columns={'name':'category_name'}, inplace=True)
#Make sure to use axis=1 and inplace=True to drop columns.
merged.drop(['id_x', 'id_y'], axis=1, inplace=True)

merged.head(5)
```

### Vraag 3b

```python
# voeg hier de code om vraag 3b te beantwoorden toe.
wijnen_num = wijnen.copy()
wijnen_num.drop(['Wijnklasse'], axis=1, inplace=True)

wijnen_num = wijnen_num - wijnen_num.min()
wijnen_num = wijnen_num / wijnen_num.max()

# wijnen_num = wijnen_num - wijnen_num.mean()
# wijnen_num = wijnen_num / wijnen_num.std()

wijnen_num.describe().transpose()
```

**Beantwoord in deze cel de gestelde vraag**

### Vraag 3c

```python
# voeg hier de code om vraag 3c te beantwoorden toe.
n_clusters_to_test = range(1, 7)
inertias = []

for n in n_clusters_to_test:
    model = KMeans(n_clusters=n).fit(wijnen_num)
    inertias.append(model.inertia_)

sns.lineplot(x=n_clusters_to_test, y=inertias)
```

**Beantwoord in deze cel de gestelde vraag**\
We gaan kiezen
voor 3 clusters want we zien in de elbow chart dat daar de grootste verandering zit.

### Vraag 3d

```python
# voeg hier de code om vraag 3d te beantwoorden toe.
model = KMeans(n_clusters=3).fit(wijnen_num)
wijnen['clusters'] = model.labels_
wijnen['clusters'] = wijnen['clusters'].astype('category')
sns.pairplot(wijnen, hue='clusters')
```

### Vraag 3e

```python
# voeg hier de code om vraag 3e te beantwoorden toe.
wijnen.groupby(['clusters']).mean()
```
